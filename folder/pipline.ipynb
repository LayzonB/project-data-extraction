```python
import pandas as pd
import numpy as np
```

```python

def rename_columns(df):
    df_rename = df.copy()
    col_n=[]
    
    col= list(df_rename.columns)
    for x in range(len(col)):
        if col[x].startswith("ga:"):
            k= col[x].replace("ga:","").upper()
            col_n.append(k)
    df_rename.columns= col_n
    return df_rename

def remove_columns(df):
    df_remove = df.copy()
    parameter_erase = len(df_remove)*0.8
    df_remove2 = df_remove == 0
    df_remove2 = df_remove2.sum()
    drop_cols = list(df_remove2[df_remove2 > parameter_erase].index)
    df_remove = df_remove.drop(drop_cols, axis=1)
    return df_remove





def final_shape(df):
    df_clean = df.copy()
    return df_clean

def prepare_auto_df(path_import,path_xport):
    df = pd.read_csv(str(path_import), index_col = 0)
    print('Dataset loaded')

    f_rename = rename_columns(df)
    print('Columnas renombradas')

    f_remove = remove_columns(f_rename)
    print('Removidas columnas sin datos')

    df_clean = final_shape(f_remove)
    print('Final dataset created')

    df_clean.to_csv(str(path_xport), index=False)
    return df_clean
```

```python
my_clean_df = prepare_auto_df("file1.csv","clean1.csv")
```
